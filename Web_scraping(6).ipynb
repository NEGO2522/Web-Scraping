## Extracting text from the tags in web page using python

```python
import requests
from bs4 import BeautifulSoup
```

```python
web = requests.get("https://www.tutorialsfreak.com")
# web
```

```python
soup = BeautifulSoup(web.content,"html.parser")
# print(soup.prettify())
```

```python
lines = soup.find_all("p")
for l in lines:
    print(l.text)
```

```python
di = soup.find_all("div")
for div in di:
    print(div.text)
```

```python
s = soup.find("div", class_= "why-choose-right-section text-center text-lg-start")
```

```python
s
```

```python
lines_1 = s.find_all("p")
```

```python
lines_1
```

```python
for li in lines_1:
    print(li.text)
```

```python
lines_1 = s.find_all("h4")
lines_1
```

```python
for h in lines_1:
    print(h.text)
```
